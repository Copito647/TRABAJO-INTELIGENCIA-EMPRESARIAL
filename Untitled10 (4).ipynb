```python
# ============================================================
# Proyecto: Credit Scoring - Give Me Some Credit
# Carga inicial del dataset
# ============================================================

from google.colab import files
import pandas as pd
import os

FIXED_NAME = 'credit_data.xlsx'

# 1. Subir el archivo (solo si no existe ya en la sesión)
if not os.path.exists(FIXED_NAME):
    print("Selecciona el archivo give_me_some_credit.xlsx desde tu ordenador")
    uploaded = files.upload()
    original_name = list(uploaded.keys())[0]
    if original_name != FIXED_NAME:
        os.rename(original_name, FIXED_NAME)
    print(f"\n✅ Archivo guardado como: {FIXED_NAME}")
    print(f"   Tamaño: {os.path.getsize(FIXED_NAME) / (1024*1024):.2f} MB\n")
else:
    print(f"✅ Usando archivo ya existente: {FIXED_NAME}\n")

# 2. Cargar los datos en el DataFrame `df`
df = pd.read_excel(FIXED_NAME)

print(f"📊 Dataset cargado correctamente")
print(f"   Filas:    {len(df):,}")
print(f"   Columnas: {len(df.columns)}")
print(f"\nPrimeras 5 filas:")
df.head()
```

```python
# ============================================================
# CELDA 1 — Imports y configuración
# ============================================================
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score,
    roc_curve, precision_recall_curve, average_precision_score,
    silhouette_score
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Estética
sns.set_style('whitegrid')
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.titlesize'] = 13
plt.rcParams['axes.labelsize'] = 11

RANDOM_STATE = 42

print("✅ Librerías cargadas")
```

```python
# ============================================================
# CELDA 2 — Inspección inicial
# ============================================================
print(f"Dimensiones: {df.shape[0]:,} filas × {df.shape[1]} columnas\n")
print("Tipos de datos y nulos:")
info_df = pd.DataFrame({
    'tipo': df.dtypes,
    'nulos': df.isnull().sum(),
    '%_nulos': (df.isnull().sum() / len(df) * 100).round(2)
})
print(info_df)

print("\nEstadísticas descriptivas:")
df.describe().round(2)
```

```python
# ============================================================
# CELDA 3 — Variable objetivo: desbalance de clases
# ============================================================
target = 'SeriousDlqin2yrs'

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

counts = df[target].value_counts()
axes[0].bar(['No moroso (0)', 'Moroso (1)'], counts.values,
            color=['#2ecc71', '#e74c3c'], edgecolor='black')
axes[0].set_title('Distribución absoluta de la variable objetivo')
axes[0].set_ylabel('Nº de clientes')
for i, v in enumerate(counts.values):
    axes[0].text(i, v + 1500, f'{v:,}', ha='center', fontweight='bold')

pcts = df[target].value_counts(normalize=True) * 100
axes[1].pie(pcts.values, labels=['No moroso', 'Moroso'],
            colors=['#2ecc71', '#e74c3c'],
            autopct='%1.2f%%', startangle=90,
            wedgeprops={'edgecolor': 'black'})
axes[1].set_title('Distribución porcentual')

plt.tight_layout()
plt.show()

print(f"\n⚠️  Desbalance fuerte: solo {pcts[1]:.2f}% de morosos.")
print("    Implicación: la accuracy NO es una métrica válida.")
print("    Usaremos AUC-ROC, Recall y F1.")
```

```python
# ============================================================
# CELDA 4 — Distribuciones de las variables predictoras
# ============================================================
predictors = [c for c in df.columns if c not in ['ID', target]]

fig, axes = plt.subplots(4, 3, figsize=(15, 14))
axes = axes.flatten()

for i, col in enumerate(predictors):
    axes[i].hist(df[col].dropna(), bins=50, color='#3498db',
                 edgecolor='black', alpha=0.7)
    axes[i].set_title(col, fontsize=10)
    axes[i].set_yscale('log')  # Log para ver outliers extremos

for j in range(len(predictors), len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('Distribuciones (escala logarítmica para detectar outliers)',
             fontsize=13, y=1.00)
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# CELDA 5 — Cómo se distribuye cada variable según el target
# ============================================================
fig, axes = plt.subplots(4, 3, figsize=(15, 14))
axes = axes.flatten()

for i, col in enumerate(predictors):
    data = [df[df[target] == 0][col].dropna(),
            df[df[target] == 1][col].dropna()]
    bp = axes[i].boxplot(data, labels=['No moroso', 'Moroso'],
                         patch_artist=True, showfliers=False)
    bp['boxes'][0].set_facecolor('#2ecc71')
    bp['boxes'][1].set_facecolor('#e74c3c')
    axes[i].set_title(col, fontsize=10)

for j in range(len(predictors), len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('Predictoras vs variable objetivo (sin outliers extremos)',
             fontsize=13, y=1.00)
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# CELDA 6 — Matriz de correlación
# ============================================================
plt.figure(figsize=(11, 9))
corr = df[predictors + [target]].corr()
mask = np.triu(np.ones_like(corr, dtype=bool), k=1)
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, vmin=-1, vmax=1, cbar_kws={'shrink': 0.8},
            linewidths=0.5, square=False)
plt.title('Matriz de correlación', fontsize=13)
plt.tight_layout()
plt.show()

print("\n🔍 Correlaciones más fuertes con la variable objetivo:")
target_corr = corr[target].drop(target).abs().sort_values(ascending=False)
print(target_corr.round(3).to_string())
```

```python
# ============================================================
# CELDA 7 — ETL: limpieza, tratamiento de outliers, imputación
# ============================================================
df_clean = df.copy()
print(f"Inicio: {df_clean.shape[0]:,} filas\n")

# 1. Edad mínima de 18 años
n_age = (df_clean['age'] < 18).sum()
df_clean = df_clean[df_clean['age'] >= 18]
print(f"✓ Filas con edad < 18 eliminadas: {n_age}")

# 2. Outliers en variables de retraso (96 y 98 son códigos de "desconocido")
late_cols = [
    'NumberOfTime30-59DaysPastDueNotWorse',
    'NumberOfTimes90DaysLate',
    'NumberOfTime60-89DaysPastDueNotWorse'
]
for col in late_cols:
    n_out = ((df_clean[col] == 96) | (df_clean[col] == 98)).sum()
    median_val = df_clean.loc[
        ~df_clean[col].isin([96, 98]), col
    ].median()
    df_clean.loc[df_clean[col].isin([96, 98]), col] = median_val
    print(f"✓ {col}: {n_out} outliers (96/98) reemplazados por mediana ({median_val})")

# 3. Imputación de nulos
n_inc = df_clean['MonthlyIncome'].isnull().sum()
df_clean['MonthlyIncome'] = df_clean['MonthlyIncome'].fillna(
    df_clean['MonthlyIncome'].median()
)
print(f"✓ MonthlyIncome: {n_inc:,} nulos imputados con mediana")

n_dep = df_clean['NumberOfDependents'].isnull().sum()
df_clean['NumberOfDependents'] = df_clean['NumberOfDependents'].fillna(0)
print(f"✓ NumberOfDependents: {n_dep:,} nulos imputados con 0")

# 4. Winsorización de RevolvingUtilization y DebtRatio (cap al percentil 99)
for col in ['RevolvingUtilizationOfUnsecuredLines', 'DebtRatio']:
    p99 = df_clean[col].quantile(0.99)
    n_cap = (df_clean[col] > p99).sum()
    df_clean[col] = df_clean[col].clip(upper=p99)
    print(f"✓ {col}: {n_cap:,} valores capados al p99 ({p99:.2f})")

print(f"\nFinal: {df_clean.shape[0]:,} filas (sin nulos: {df_clean.isnull().sum().sum() == 0})")
```

```python
# ============================================================
# CELDA 8 — Creación de variables derivadas
# ============================================================
df_clean['IncomePerDependent'] = df_clean['MonthlyIncome'] / (df_clean['NumberOfDependents'] + 1)
df_clean['TotalLatePayments'] = (
    df_clean['NumberOfTime30-59DaysPastDueNotWorse'] +
    df_clean['NumberOfTime60-89DaysPastDueNotWorse'] +
    df_clean['NumberOfTimes90DaysLate']
)
df_clean['HasLatePayments'] = (df_clean['TotalLatePayments'] > 0).astype(int)
df_clean['AgeGroup'] = pd.cut(
    df_clean['age'],
    bins=[0, 30, 45, 60, 120],
    labels=['Joven (<30)', 'Adulto (30-45)', 'Maduro (45-60)', 'Senior (>60)']
)

print("✅ Variables derivadas creadas:")
print("   • IncomePerDependent  — ingreso per cápita del hogar")
print("   • TotalLatePayments   — suma total de retrasos históricos")
print("   • HasLatePayments     — flag binario de historial de retrasos")
print("   • AgeGroup            — segmentación por edad")

print("\nTasa de morosidad por grupo de edad:")
print((df_clean.groupby('AgeGroup')[target].mean() * 100).round(2).astype(str) + ' %')
```

```python
# ============================================================
# CELDA 9 — Clustering K-Means para descubrir perfiles
# ============================================================
cluster_features = [
    'age', 'MonthlyIncome', 'DebtRatio',
    'RevolvingUtilizationOfUnsecuredLines',
    'NumberOfOpenCreditLinesAndLoans', 'TotalLatePayments'
]

X_cluster = df_clean[cluster_features].copy()
scaler_c = StandardScaler()
X_cluster_scaled = scaler_c.fit_transform(X_cluster)

# Método del codo + silhouette para elegir K
print("Evaluando K óptimo (esto tarda ~1 min)...")
inertias, silhouettes = [], []
K_range = range(2, 8)
for k in K_range:
    km = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10)
    labels = km.fit_predict(X_cluster_scaled)
    inertias.append(km.inertia_)
    sample = np.random.choice(len(X_cluster_scaled), 10000, replace=False)
    silhouettes.append(silhouette_score(X_cluster_scaled[sample], labels[sample]))

fig, axes = plt.subplots(1, 2, figsize=(13, 4))
axes[0].plot(K_range, inertias, 'o-', color='#3498db', linewidth=2)
axes[0].set_xlabel('Número de clusters (K)')
axes[0].set_ylabel('Inercia')
axes[0].set_title('Método del codo')

axes[1].plot(K_range, silhouettes, 'o-', color='#e67e22', linewidth=2)
axes[1].set_xlabel('Número de clusters (K)')
axes[1].set_ylabel('Silhouette score')
axes[1].set_title('Coeficiente de silueta')

plt.tight_layout()
plt.show()

# Aplicar K-Means con K=4 (justifícalo en el informe según los gráficos)
K_OPTIMAL = 4
kmeans = KMeans(n_clusters=K_OPTIMAL, random_state=RANDOM_STATE, n_init=10)
df_clean['Cluster'] = kmeans.fit_predict(X_cluster_scaled)
print(f"\n✅ {K_OPTIMAL} clusters asignados")
```

```python
# ============================================================
# CELDA 10 — Perfilado de los clusters
# ============================================================
print("Características medias por cluster:\n")
profile = df_clean.groupby('Cluster')[cluster_features + [target]].mean().round(2)
profile['N_clientes'] = df_clean.groupby('Cluster').size()
profile['Tasa_morosidad_%'] = (df_clean.groupby('Cluster')[target].mean() * 100).round(2)
print(profile.to_string())

# Visualización de la tasa de morosidad por cluster
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

cluster_size = df_clean['Cluster'].value_counts().sort_index()
axes[0].bar(cluster_size.index, cluster_size.values,
            color='#3498db', edgecolor='black')
axes[0].set_xlabel('Cluster')
axes[0].set_ylabel('Nº de clientes')
axes[0].set_title('Tamaño de cada cluster')
for i, v in enumerate(cluster_size.values):
    axes[0].text(i, v + 1000, f'{v:,}', ha='center', fontweight='bold')

cluster_risk = df_clean.groupby('Cluster')[target].mean() * 100
colors = ['#e74c3c' if r > cluster_risk.mean() else '#2ecc71'
          for r in cluster_risk.values]
axes[1].bar(cluster_risk.index, cluster_risk.values,
            color=colors, edgecolor='black')
axes[1].axhline(cluster_risk.mean(), color='black',
                linestyle='--', label=f'Media: {cluster_risk.mean():.1f}%')
axes[1].set_xlabel('Cluster')
axes[1].set_ylabel('% morosidad')
axes[1].set_title('Tasa de morosidad por cluster')
axes[1].legend()
for i, v in enumerate(cluster_risk.values):
    axes[1].text(i, v + 0.3, f'{v:.1f}%', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# ============================================================
# CELDA 11 — Mapa de clusters en 2D usando PCA
# ============================================================
pca = PCA(n_components=2, random_state=RANDOM_STATE)
X_pca = pca.fit_transform(X_cluster_scaled)

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Mapa coloreado por cluster
sample_idx = np.random.choice(len(X_pca), 8000, replace=False)
scatter1 = axes[0].scatter(X_pca[sample_idx, 0], X_pca[sample_idx, 1],
                            c=df_clean['Cluster'].values[sample_idx],
                            cmap='tab10', alpha=0.5, s=15)
axes[0].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% var.)')
axes[0].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% var.)')
axes[0].set_title('Mapa de clusters (proyección PCA 2D)')
plt.colorbar(scatter1, ax=axes[0], label='Cluster')

# Mapa coloreado por morosidad
colors_risk = df_clean[target].values[sample_idx]
scatter2 = axes[1].scatter(X_pca[sample_idx, 0], X_pca[sample_idx, 1],
                            c=colors_risk, cmap='RdYlGn_r',
                            alpha=0.5, s=15)
axes[1].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% var.)')
axes[1].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% var.)')
axes[1].set_title('Mismo mapa coloreado por morosidad real')
plt.colorbar(scatter2, ax=axes[1], label='Moroso (1) / No moroso (0)')

plt.tight_layout()
plt.show()

print(f"Varianza explicada por las 2 componentes: "
      f"{pca.explained_variance_ratio_.sum()*100:.1f}%")
```

```python
# ============================================================
# CELDA 12 — Train/test split
# ============================================================
features_model = [
    'RevolvingUtilizationOfUnsecuredLines', 'age',
    'NumberOfTime30-59DaysPastDueNotWorse', 'DebtRatio', 'MonthlyIncome',
    'NumberOfOpenCreditLinesAndLoans', 'NumberOfTimes90DaysLate',
    'NumberRealEstateLoansOrLines', 'NumberOfTime60-89DaysPastDueNotWorse',
    'NumberOfDependents', 'IncomePerDependent', 'TotalLatePayments',
    'HasLatePayments'
]

X = df_clean[features_model]
y = df_clean[target]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=RANDOM_STATE, stratify=y
)

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)

print(f"Train: {X_train.shape[0]:,} filas")
print(f"Test:  {X_test.shape[0]:,} filas")
print(f"Tasa morosidad train: {y_train.mean()*100:.2f}%")
print(f"Tasa morosidad test:  {y_test.mean()*100:.2f}%")
```

```python
# ============================================================
# CELDA 13 — Entrenamiento de modelos
# ============================================================
models = {
    'Regresión Logística': LogisticRegression(
        class_weight='balanced', max_iter=1000, random_state=RANDOM_STATE
    ),
    'Random Forest': RandomForestClassifier(
        n_estimators=150, max_depth=10, class_weight='balanced',
        random_state=RANDOM_STATE, n_jobs=-1
    ),
    'Gradient Boosting': GradientBoostingClassifier(
        n_estimators=150, max_depth=5, random_state=RANDOM_STATE
    )
}

results = {}
for name, model in models.items():
    print(f"⏳ Entrenando {name}...")
    if name == 'Regresión Logística':
        model.fit(X_train_sc, y_train)
        y_pred = model.predict(X_test_sc)
        y_proba = model.predict_proba(X_test_sc)[:, 1]
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:, 1]

    results[name] = {
        'model': model,
        'y_pred': y_pred,
        'y_proba': y_proba,
        'auc': roc_auc_score(y_test, y_proba),
        'avg_precision': average_precision_score(y_test, y_proba)
    }
    print(f"   ✓ AUC = {results[name]['auc']:.4f}")

print("\n📊 Resumen comparativo:")
summary = pd.DataFrame({
    'Modelo': list(results.keys()),
    'AUC-ROC': [results[m]['auc'] for m in results],
    'Avg Precision': [results[m]['avg_precision'] for m in results]
}).round(4)
print(summary.to_string(index=False))
```

```python
# ============================================================
# CELDA 14 — Curvas ROC y PR
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for name, res in results.items():
    fpr, tpr, _ = roc_curve(y_test, res['y_proba'])
    axes[0].plot(fpr, tpr, label=f"{name} (AUC = {res['auc']:.3f})", linewidth=2)

axes[0].plot([0, 1], [0, 1], 'k--', alpha=0.5)
axes[0].set_xlabel('Tasa de falsos positivos')
axes[0].set_ylabel('Tasa de verdaderos positivos')
axes[0].set_title('Curva ROC')
axes[0].legend(loc='lower right')

for name, res in results.items():
    prec, rec, _ = precision_recall_curve(y_test, res['y_proba'])
    axes[1].plot(rec, prec,
                 label=f"{name} (AP = {res['avg_precision']:.3f})", linewidth=2)

axes[1].axhline(y_test.mean(), color='gray', linestyle='--',
                alpha=0.5, label='Baseline (clase positiva)')
axes[1].set_xlabel('Recall')
axes[1].set_ylabel('Precision')
axes[1].set_title('Curva Precision-Recall')
axes[1].legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# ============================================================
# CELDA 15 — Matriz de confusión + classification report
# ============================================================
best_name = max(results, key=lambda k: results[k]['auc'])
best = results[best_name]
print(f"🏆 Mejor modelo: {best_name} (AUC = {best['auc']:.4f})\n")

cm = confusion_matrix(y_test, best['y_pred'])

fig, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt=',d', cmap='Blues', cbar=False,
            xticklabels=['No moroso', 'Moroso'],
            yticklabels=['No moroso', 'Moroso'], ax=ax,
            annot_kws={'size': 14, 'weight': 'bold'})
ax.set_xlabel('Predicción', fontsize=11)
ax.set_ylabel('Real', fontsize=11)
ax.set_title(f'Matriz de confusión — {best_name}', fontsize=13)
plt.tight_layout()
plt.show()

print("\n📋 Classification report:\n")
print(classification_report(y_test, best['y_pred'],
                            target_names=['No moroso', 'Moroso']))
```

```python
# ============================================================
# CELDA 16 — Importancia de variables del mejor modelo
# ============================================================
if hasattr(best['model'], 'feature_importances_'):
    importances = pd.DataFrame({
        'Variable': features_model,
        'Importancia': best['model'].feature_importances_
    }).sort_values('Importancia', ascending=True)

    plt.figure(figsize=(10, 6))
    plt.barh(importances['Variable'], importances['Importancia'],
             color='#3498db', edgecolor='black')
    plt.xlabel('Importancia')
    plt.title(f'Importancia de variables — {best_name}', fontsize=13)
    plt.tight_layout()
    plt.show()

    print("\n🔝 Top 5 variables más importantes:")
    print(importances.tail(5).iloc[::-1].to_string(index=False))
else:
    coefs = pd.DataFrame({
        'Variable': features_model,
        'Coeficiente': best['model'].coef_[0]
    }).sort_values('Coeficiente', key=abs, ascending=True)

    plt.figure(figsize=(10, 6))
    colors = ['#e74c3c' if c > 0 else '#2ecc71'
              for c in coefs['Coeficiente']]
    plt.barh(coefs['Variable'], coefs['Coeficiente'],
             color=colors, edgecolor='black')
    plt.axvline(0, color='black', linewidth=0.8)
    plt.xlabel('Coeficiente')
    plt.title(f'Coeficientes — {best_name}', fontsize=13)
    plt.tight_layout()
    plt.show()
```

```python
# ============================================================
# CELDA 17 — Segmentación de clientes por nivel de riesgo
# ============================================================
risk_df = pd.DataFrame({
    'real': y_test.values,
    'probabilidad_default': best['y_proba']
})
risk_df['Segmento_riesgo'] = pd.cut(
    risk_df['probabilidad_default'],
    bins=[0, 0.1, 0.3, 0.6, 1.0],
    labels=['Bajo', 'Medio', 'Alto', 'Muy alto']
)

segment_summary = risk_df.groupby('Segmento_riesgo').agg(
    N_clientes=('real', 'count'),
    Tasa_morosidad=('real', 'mean')
).round(4)
segment_summary['Tasa_morosidad'] = (segment_summary['Tasa_morosidad'] * 100).round(2).astype(str) + ' %'
print("📊 Segmentación final por nivel de riesgo:\n")
print(segment_summary.to_string())

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

seg_counts = risk_df['Segmento_riesgo'].value_counts().sort_index()
axes[0].bar(seg_counts.index.astype(str), seg_counts.values,
            color=['#2ecc71', '#f39c12', '#e67e22', '#e74c3c'],
            edgecolor='black')
axes[0].set_ylabel('Nº de clientes')
axes[0].set_title('Nº de clientes por segmento')
for i, v in enumerate(seg_counts.values):
    axes[0].text(i, v + 200, f'{v:,}', ha='center', fontweight='bold')

seg_real = risk_df.groupby('Segmento_riesgo')['real'].mean() * 100
axes[1].bar(seg_real.index.astype(str), seg_real.values,
            color=['#2ecc71', '#f39c12', '#e67e22', '#e74c3c'],
            edgecolor='black')
axes[1].set_ylabel('% morosidad real')
axes[1].set_title('Tasa real de morosidad por segmento')
for i, v in enumerate(seg_real.values):
    axes[1].text(i, v + 0.5, f'{v:.1f}%', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# ╔══════════════════════════════════════════════════════════════════╗
# ║  MEJORAS POST-FEEDBACK — Proyecto Credit Scoring                 ║
# ║                                                                  ║
# ║  Pegar al final del notebook (después del setup pro).            ║
# ║  Cada bloque entre líneas '#===' es UNA celda del notebook.      ║
# ║                                                                  ║
# ║  Requiere ejecutadas previamente las CELDAS 0-17 originales:     ║
# ║  df_clean, target, features_model, X_train, X_test, y_train,     ║
# ║  y_test, X_train_sc, X_test_sc, models, results, best, best_name ║
# ╚══════════════════════════════════════════════════════════════════╝


# ============================================================
# CELDA FB-0 — Setup (paleta + carpeta de salida)
# ============================================================
# Si ya ejecutaste el setup de los gráficos pro, esto es redundante;
# lo dejo standalone por si lo ejecutas en una sesión nueva.

!pip install -q shap kaleido

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.metrics import (roc_auc_score, confusion_matrix, classification_report,
                              precision_recall_curve, recall_score, precision_score,
                              f1_score)
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler

COLOR_BG    = '#0E1117'
COLOR_PANEL = '#161A23'
COLOR_GRID  = '#2A2F3A'
COLOR_TEXT  = '#FAFAFA'

OUT_FB = 'graficos_feedback'
os.makedirs(OUT_FB, exist_ok=True)

plt.rcParams.update({
    'figure.facecolor':  COLOR_BG,
    'axes.facecolor':    COLOR_PANEL,
    'axes.edgecolor':    COLOR_GRID,
    'axes.labelcolor':   COLOR_TEXT,
    'text.color':        COLOR_TEXT,
    'xtick.color':       COLOR_TEXT,
    'ytick.color':       COLOR_TEXT,
    'grid.color':        COLOR_GRID,
    'grid.alpha':        0.4,
    'axes.titleweight':  'bold',
    'axes.titlesize':    14,
    'axes.titlecolor':   COLOR_TEXT,
})

def save_fb(name):
    plt.savefig(f'{OUT_FB}/{name}.png', dpi=200, bbox_inches='tight',
                facecolor=COLOR_BG, edgecolor='none')

print('✅ Setup feedback listo. Carpeta:', OUT_FB)


# ============================================================
# CELDA FB-1 — Validación cruzada k=5 estratificada
# (B.3 del roadmap — esfuerzo bajo, impacto medio)
# ============================================================
print('🔄 Calculando validación cruzada k=5 estratificada...\n')

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_results = {}

# Recombinar X completo (train + test) para hacer CV honesta
X_full = pd.concat([X_train, X_test])
y_full = pd.concat([y_train, y_test])

for name, model in models.items():
    if name == 'Regresión Logística':
        # Regresión logística necesita escalado: lo manejamos en pipeline
        from sklearn.pipeline import Pipeline
        from sklearn.linear_model import LogisticRegression
        pipe = Pipeline([
            ('scaler', StandardScaler()),
            ('clf', LogisticRegression(max_iter=1000, class_weight='balanced',
                                        random_state=42))
        ])
        scores = cross_val_score(pipe, X_full, y_full, cv=cv,
                                 scoring='roc_auc', n_jobs=-1)
    else:
        scores = cross_val_score(model, X_full, y_full, cv=cv,
                                 scoring='roc_auc', n_jobs=-1)
    cv_results[name] = scores
    print(f'  {name:25s}: AUC = {scores.mean():.3f} ± {scores.std():.3f}  '
          f'[{scores.min():.3f}, {scores.max():.3f}]')

# Visualización del CV
fig, ax = plt.subplots(figsize=(11, 5))
positions = np.arange(len(cv_results))
data = [cv_results[n] for n in cv_results]
labels = list(cv_results.keys())

bp = ax.boxplot(data, positions=positions, patch_artist=True, widths=0.5,
                medianprops=dict(color='#FFD166', linewidth=2),
                boxprops=dict(facecolor='#161A23', edgecolor='#06D6A0', linewidth=1.5),
                whiskerprops=dict(color='#06D6A0', linewidth=1.5),
                capprops=dict(color='#06D6A0', linewidth=1.5),
                flierprops=dict(marker='o', markerfacecolor='#EF476F',
                                markersize=6, markeredgecolor='white'))

# Puntos individuales encima
for i, (name, scores) in enumerate(cv_results.items()):
    ax.scatter([i] * len(scores), scores, color='#06D6A0', s=40, zorder=3,
               edgecolor='white', linewidth=1, alpha=0.8)
    # Label con media ± std
    ax.annotate(f'{scores.mean():.3f}±{scores.std():.3f}',
                xy=(i, scores.max()), xytext=(0, 10),
                textcoords='offset points', ha='center',
                color=COLOR_TEXT, fontsize=10, weight='bold')

ax.set_xticks(positions)
ax.set_xticklabels(labels, size=11)
ax.set_ylabel('AUC-ROC', size=12)
ax.set_title('Validación cruzada estratificada k=5 — AUC por modelo y fold',
             size=13)
ax.grid(axis='y', alpha=0.3, color=COLOR_GRID)
for s in ['top', 'right']: ax.spines[s].set_visible(False)
plt.tight_layout()
save_fb('FB1_cv_kfold')
plt.show()

# Guardar también el resumen para el Word
cv_summary = pd.DataFrame({
    name: [scores.mean(), scores.std(), scores.min(), scores.max()]
    for name, scores in cv_results.items()
}, index=['AUC mean', 'AUC std', 'AUC min', 'AUC max']).T.round(4)
print('\nResumen CV:')
print(cv_summary)


# ============================================================
# CELDA FB-2 — Optimización de umbral con matriz de coste
# (B.1 del roadmap — esfuerzo medio, IMPACTO ALTO)
# ============================================================
print('🔄 Optimizando umbral con matriz de coste...\n')

# Suponemos que un FN cuesta 10 veces más que un FP (consensuable con Riesgos)
COST_FN = 10  # coste de un moroso no detectado
COST_FP = 1   # coste de un cliente solvente revisado innecesariamente

y_proba = best['y_proba']
thresholds = np.linspace(0.01, 0.99, 99)

cost_per_threshold = []
metrics_per_threshold = []

for thr in thresholds:
    y_pred = (y_proba >= thr).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    cost = fn * COST_FN + fp * COST_FP
    cost_per_threshold.append(cost)
    rec = tp / (tp + fn) if (tp + fn) > 0 else 0
    prec = tp / (tp + fp) if (tp + fp) > 0 else 0
    f1 = 2 * prec * rec / (prec + rec) if (prec + rec) > 0 else 0
    metrics_per_threshold.append({'thr': thr, 'tn': tn, 'fp': fp, 'fn': fn, 'tp': tp,
                                   'recall': rec, 'precision': prec, 'f1': f1, 'cost': cost})

df_metrics = pd.DataFrame(metrics_per_threshold)

# Umbral óptimo
opt_idx = df_metrics['cost'].idxmin()
opt_thr = df_metrics.loc[opt_idx, 'thr']
opt_recall = df_metrics.loc[opt_idx, 'recall']
opt_precision = df_metrics.loc[opt_idx, 'precision']
opt_f1 = df_metrics.loc[opt_idx, 'f1']
opt_cost = df_metrics.loc[opt_idx, 'cost']

print(f'  Coste FN:FP asumido = {COST_FN}:{COST_FP}')
print(f'  Umbral óptimo       = {opt_thr:.3f}')
print(f'  Recall en óptimo    = {opt_recall*100:.1f}%')
print(f'  Precision en óptimo = {opt_precision*100:.1f}%')
print(f'  F1 en óptimo        = {opt_f1:.3f}')
print(f'  Coste total mínimo  = {opt_cost:,.0f}')

# Comparación con umbral 0.5
m05 = df_metrics[df_metrics['thr'].round(2) == 0.50].iloc[0]
print(f'\n  vs. umbral por defecto (0.5):')
print(f'    Recall {m05["recall"]*100:.1f}% → {opt_recall*100:.1f}% '
      f'(+{(opt_recall-m05["recall"])*100:.1f} puntos)')
print(f'    Coste {m05["cost"]:,.0f} → {opt_cost:,.0f} '
      f'({(opt_cost-m05["cost"])/m05["cost"]*100:+.1f}%)')

# Visualización
fig, axes = plt.subplots(1, 2, figsize=(14, 5.5))

# Coste por umbral
ax = axes[0]
ax.plot(df_metrics['thr'], df_metrics['cost'], color='#FFD166', linewidth=2.5)
ax.fill_between(df_metrics['thr'], 0, df_metrics['cost'], color='#FFD166', alpha=0.15)
ax.axvline(opt_thr, color='#06D6A0', linestyle='--', linewidth=2,
           label=f'Umbral óptimo = {opt_thr:.2f}')
ax.axvline(0.5, color='#EF476F', linestyle=':', linewidth=1.5, alpha=0.7,
           label='Umbral por defecto = 0.50')
ax.scatter([opt_thr], [opt_cost], s=150, color='#06D6A0', zorder=5,
           edgecolor='white', linewidths=2)
ax.set_xlabel('Umbral de clasificación', size=12)
ax.set_ylabel(f'Coste total (FN×{COST_FN} + FP×{COST_FP})', size=12)
ax.set_title(f'Coste por umbral — relación FN:FP = {COST_FN}:{COST_FP}', size=13)
ax.grid(alpha=0.3, color=COLOR_GRID)
ax.legend(facecolor=COLOR_PANEL, edgecolor=COLOR_GRID, labelcolor=COLOR_TEXT)
for s in ['top', 'right']: ax.spines[s].set_visible(False)

# Recall vs Precision por umbral
ax = axes[1]
ax.plot(df_metrics['thr'], df_metrics['recall'], color='#06D6A0',
        linewidth=2.5, label='Recall')
ax.plot(df_metrics['thr'], df_metrics['precision'], color='#118AB2',
        linewidth=2.5, label='Precision')
ax.plot(df_metrics['thr'], df_metrics['f1'], color='#FFD166',
        linewidth=2, label='F1 score', linestyle='--')
ax.axvline(opt_thr, color='#06D6A0', linestyle='--', linewidth=2, alpha=0.7)
ax.axvline(0.5, color='#EF476F', linestyle=':', linewidth=1.5, alpha=0.5)
ax.set_xlabel('Umbral de clasificación', size=12)
ax.set_ylabel('Métrica', size=12)
ax.set_title('Recall, Precision y F1 por umbral', size=13)
ax.set_ylim(0, 1.05)
ax.grid(alpha=0.3, color=COLOR_GRID)
ax.legend(facecolor=COLOR_PANEL, edgecolor=COLOR_GRID, labelcolor=COLOR_TEXT)
for s in ['top', 'right']: ax.spines[s].set_visible(False)

plt.suptitle(f'Optimización de umbral — Gradient Boosting',
             color=COLOR_TEXT, size=15, weight='bold', y=1.02)
plt.tight_layout()
save_fb('FB2_threshold_optimo')
plt.show()

# Matriz de confusión con el umbral óptimo
import seaborn as sns
y_pred_opt = (y_proba >= opt_thr).astype(int)
cm_opt = confusion_matrix(y_test, y_pred_opt)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
for ax, cm, title, thr in zip(axes,
                              [confusion_matrix(y_test, (y_proba >= 0.5).astype(int)), cm_opt],
                              ['Umbral 0.5 (por defecto)', f'Umbral {opt_thr:.2f} (óptimo)'],
                              [0.5, opt_thr]):
    sns.heatmap(cm, annot=True, fmt=',d', cmap='Blues', ax=ax,
                xticklabels=['No moroso', 'Moroso'],
                yticklabels=['No moroso', 'Moroso'],
                cbar=False, annot_kws={'size': 16, 'weight': 'bold'},
                linewidths=2, linecolor=COLOR_BG)
    ax.set_title(title, size=13)
    ax.set_xlabel('Predicción', size=11)
    ax.set_ylabel('Real', size=11)

plt.suptitle('Comparativa de matrices de confusión',
             color=COLOR_TEXT, size=15, weight='bold')
plt.tight_layout()
save_fb('FB2b_confusion_comparativa')
plt.show()


# ============================================================
# CELDA FB-3 — Análisis de fairness por grupo de edad
# (B.4 del roadmap — esfuerzo bajo, impacto alto)
# ============================================================
print('🔄 Calculando métricas de equidad por grupo de edad...\n')

# Reconstruir AgeGroup en el test
risk_full = df_clean.loc[X_test.index].copy()
risk_full['y_pred_05']  = (y_proba >= 0.5).astype(int)
risk_full['y_pred_opt'] = (y_proba >= opt_thr).astype(int)
risk_full['y_real']     = y_test.values
risk_full['y_proba']    = y_proba

fairness_rows = []
for ag in risk_full['AgeGroup'].cat.categories:
    sub = risk_full[risk_full['AgeGroup'] == ag]
    n = len(sub)
    pos = sub['y_real'].sum()
    if n == 0 or pos == 0:
        continue
    auc_g = roc_auc_score(sub['y_real'], sub['y_proba'])
    # En umbral por defecto
    rec_05 = recall_score(sub['y_real'], sub['y_pred_05'], zero_division=0)
    pre_05 = precision_score(sub['y_real'], sub['y_pred_05'], zero_division=0)
    # En umbral óptimo
    rec_opt = recall_score(sub['y_real'], sub['y_pred_opt'], zero_division=0)
    pre_opt = precision_score(sub['y_real'], sub['y_pred_opt'], zero_division=0)
    fairness_rows.append({
        'AgeGroup': str(ag), 'n': n, 'morosos_reales': int(pos),
        'AUC': auc_g,
        'Recall@0.5': rec_05, 'Precision@0.5': pre_05,
        f'Recall@{opt_thr:.2f}': rec_opt,
        f'Precision@{opt_thr:.2f}': pre_opt,
    })
df_fair = pd.DataFrame(fairness_rows)
print(df_fair.round(3).to_string(index=False))

# Visualización
fig, axes = plt.subplots(1, 2, figsize=(15, 5.5))

# AUC por grupo
ax = axes[0]
colors_age = ['#EF476F', '#FF8C42', '#FFD166', '#06D6A0']
bars = ax.bar(df_fair['AgeGroup'], df_fair['AUC'],
              color=colors_age[:len(df_fair)], edgecolor='white', linewidth=2,
              alpha=0.85)
for bar, val in zip(bars, df_fair['AUC']):
    ax.text(bar.get_x() + bar.get_width()/2, val + 0.01,
            f'{val:.3f}', ha='center', va='bottom',
            color=COLOR_TEXT, fontweight='bold', fontsize=11)
ax.axhline(roc_auc_score(y_test, y_proba), color='#FFD166', linestyle='--',
           linewidth=1.5, alpha=0.7,
           label=f'AUC global = {roc_auc_score(y_test, y_proba):.3f}')
ax.set_ylabel('AUC-ROC', size=12)
ax.set_title('AUC por grupo de edad — ¿el modelo discrimina igual de bien?', size=12)
ax.set_ylim(0, 1.05)
ax.legend(facecolor=COLOR_PANEL, edgecolor=COLOR_GRID, labelcolor=COLOR_TEXT)
ax.grid(axis='y', alpha=0.3, color=COLOR_GRID)
for s in ['top', 'right']: ax.spines[s].set_visible(False)

# Recall por grupo
ax = axes[1]
x = np.arange(len(df_fair))
width = 0.35
ax.bar(x - width/2, df_fair['Recall@0.5'], width, color='#EF476F',
       edgecolor='white', linewidth=1.5, label='Recall @ 0.50', alpha=0.85)
ax.bar(x + width/2, df_fair[f'Recall@{opt_thr:.2f}'], width, color='#06D6A0',
       edgecolor='white', linewidth=1.5,
       label=f'Recall @ {opt_thr:.2f} (óptimo)', alpha=0.85)
for i, (r05, ropt) in enumerate(zip(df_fair['Recall@0.5'],
                                     df_fair[f'Recall@{opt_thr:.2f}'])):
    ax.text(i - width/2, r05 + 0.02, f'{r05*100:.0f}%',
            ha='center', color=COLOR_TEXT, fontsize=9, weight='bold')
    ax.text(i + width/2, ropt + 0.02, f'{ropt*100:.0f}%',
            ha='center', color=COLOR_TEXT, fontsize=9, weight='bold')
ax.set_xticks(x)
ax.set_xticklabels(df_fair['AgeGroup'])
ax.set_ylabel('Recall (% morosos detectados)', size=12)
ax.set_title('Recall por grupo de edad — equidad en la detección', size=12)
ax.set_ylim(0, 1.05)
ax.legend(facecolor=COLOR_PANEL, edgecolor=COLOR_GRID, labelcolor=COLOR_TEXT)
ax.grid(axis='y', alpha=0.3, color=COLOR_GRID)
for s in ['top', 'right']: ax.spines[s].set_visible(False)

plt.suptitle('Análisis de equidad (fairness) por grupo de edad',
             color=COLOR_TEXT, size=15, weight='bold', y=1.02)
plt.tight_layout()
save_fb('FB3_fairness_edad')
plt.show()


# ============================================================
# CELDA FB-4 (FIX) — SHAP corregido para versiones nuevas
# ============================================================
print('🔄 Calculando SHAP values (versión corregida)...\n')

import shap

explainer = shap.TreeExplainer(models[best_name])
sample_size = 2000
sample_idx = np.random.RandomState(42).choice(len(X_test), sample_size, replace=False)
X_sample = X_test.iloc[sample_idx]
shap_values = explainer.shap_values(X_sample)

# 🔑 FIX: normalizar shapes — versiones nuevas devuelven 3D o lista
shap_values = np.asarray(shap_values)
if shap_values.ndim == 3:                       # (n_samples, n_features, n_classes)
    shap_values = shap_values[:, :, 1]          # quedarse con clase positiva
# Si era lista [class0, class1], asarray la convirtió en 3D arriba.

# 🔑 FIX: expected_value puede ser escalar o array
expected_value = explainer.expected_value
if hasattr(expected_value, '__len__'):
    expected_value = float(np.asarray(expected_value).flatten()[-1])  # clase positiva
else:
    expected_value = float(expected_value)

print(f'  Shape de shap_values: {shap_values.shape}')
print(f'  Expected value (base): {expected_value:.3f}')

proba_sample = y_proba[sample_idx]
idx_bajo = int(np.argmin(proba_sample))
idx_alto = int(np.argmax(proba_sample))


def plot_shap_waterfall(shap_vals, base_value, feature_names, feature_values,
                       title, save_name):
    # Coerce todo a tipos seguros
    shap_vals = np.asarray(shap_vals).flatten()
    base_value = float(base_value)
    total_contrib = float(shap_vals.sum())

    contribs = list(zip(feature_names, feature_values, shap_vals))
    contribs.sort(key=lambda x: abs(x[2]), reverse=True)
    contribs = contribs[:10]

    fig, ax = plt.subplots(figsize=(12, 6))
    y_pos = np.arange(len(contribs))[::-1]
    values = [float(c[2]) for c in contribs]
    labels = [f'{c[0]} = {float(c[1]):.2f}' for c in contribs]
    colors = ['#EF476F' if v > 0 else '#06D6A0' for v in values]

    bars = ax.barh(y_pos, values, color=colors, edgecolor='white', linewidth=1.5,
                   alpha=0.85)
    for bar, val in zip(bars, values):
        sign = '+' if val > 0 else ''
        ax.text(val + (0.005 if val > 0 else -0.005),
                bar.get_y() + bar.get_height()/2,
                f'{sign}{val:.3f}',
                va='center', ha='left' if val > 0 else 'right',
                color=COLOR_TEXT, fontsize=10, weight='bold')

    ax.set_yticks(y_pos)
    ax.set_yticklabels(labels, fontsize=10)
    ax.axvline(0, color='gray', linewidth=1)
    ax.set_xlabel('Contribución SHAP (impacto en log-odds del default)', size=11)
    ax.set_title(title, size=13)
    ax.grid(axis='x', alpha=0.3, color=COLOR_GRID)
    for s in ['top', 'right']: ax.spines[s].set_visible(False)

    proba_final = 1 / (1 + np.exp(-(base_value + total_contrib)))
    ax.text(0.98, 0.02,
            f'Base value (log-odds): {base_value:.3f}\n'
            f'Suma de contribuciones: {total_contrib:+.3f}\n'
            f'Probabilidad final: {proba_final*100:.1f}%',
            transform=ax.transAxes, ha='right', va='bottom',
            color=COLOR_TEXT, fontsize=10,
            bbox=dict(boxstyle='round,pad=0.5', facecolor='#161A23',
                      edgecolor='gray', alpha=0.8))
    plt.tight_layout()
    save_fb(save_name)
    plt.show()


plot_shap_waterfall(
    shap_values[idx_bajo], expected_value,
    features_model, X_sample.iloc[idx_bajo].values,
    f'SHAP — Cliente Bajo riesgo (probabilidad de default = {proba_sample[idx_bajo]*100:.1f}%)',
    'FB4a_shap_cliente_bajo'
)

plot_shap_waterfall(
    shap_values[idx_alto], expected_value,
    features_model, X_sample.iloc[idx_alto].values,
    f'SHAP — Cliente Muy alto riesgo (probabilidad de default = {proba_sample[idx_alto]*100:.1f}%)',
    'FB4b_shap_cliente_alto'
)

# SHAP summary plot global
fig = plt.figure(figsize=(11, 6))
shap.summary_plot(shap_values, X_sample, feature_names=features_model,
                   show=False, plot_size=None, max_display=10)
plt.gcf().set_facecolor(COLOR_BG)
ax = plt.gca()
ax.set_facecolor(COLOR_PANEL)
ax.tick_params(colors=COLOR_TEXT)
for spine in ax.spines.values():
    spine.set_color(COLOR_GRID)
plt.title('SHAP summary plot — distribución global de impactos',
          color=COLOR_TEXT, size=13, weight='bold')
plt.tight_layout()
save_fb('FB4c_shap_summary')
plt.show()

print('\n✅ SHAP completado')

# ============================================================
# CELDA FB-5 — Information Value (IV) y Weight of Evidence (WoE)
# (B.5 del roadmap — esfuerzo bajo, estándar bancario)
# ============================================================
print('🔄 Calculando IV y WoE...\n')

def calculate_iv_woe(df, feature, target, bins=10):
    """Calcula IV y WoE de una variable continua mediante binning."""
    data = df[[feature, target]].copy().dropna()
    # Bins por cuantiles, manejando duplicados
    try:
        data['bin'] = pd.qcut(data[feature], q=bins, duplicates='drop')
    except Exception:
        data['bin'] = pd.cut(data[feature], bins=bins)

    grouped = data.groupby('bin', observed=True).agg(
        total=(target, 'count'),
        events=(target, 'sum'),
    )
    grouped['non_events'] = grouped['total'] - grouped['events']
    total_events = grouped['events'].sum()
    total_non_events = grouped['non_events'].sum()

    # Suavizado para evitar log(0)
    grouped['pct_events']     = (grouped['events']     + 0.5) / (total_events     + 0.5)
    grouped['pct_non_events'] = (grouped['non_events'] + 0.5) / (total_non_events + 0.5)
    grouped['woe'] = np.log(grouped['pct_non_events'] / grouped['pct_events'])
    grouped['iv']  = (grouped['pct_non_events'] - grouped['pct_events']) * grouped['woe']
    return grouped['iv'].sum(), grouped

iv_results = []
for feat in features_model:
    iv, _ = calculate_iv_woe(df_clean, feat, target, bins=10)
    iv_results.append({'variable': feat, 'IV': iv})

iv_df = pd.DataFrame(iv_results).sort_values('IV', ascending=False).reset_index(drop=True)

def iv_strength(iv):
    if iv < 0.02: return 'Inútil'
    elif iv < 0.1: return 'Débil'
    elif iv < 0.3: return 'Media'
    elif iv < 0.5: return 'Fuerte'
    else: return 'Sospechoso'

iv_df['Fuerza'] = iv_df['IV'].apply(iv_strength)
print(iv_df.round(4).to_string(index=False))

# Visualización
fig, ax = plt.subplots(figsize=(11, 7))
strength_colors = {'Inútil': '#666666', 'Débil': '#FFD166', 'Media': '#FF8C42',
                    'Fuerte': '#06D6A0', 'Sospechoso': '#EF476F'}
colors_iv = [strength_colors[s] for s in iv_df['Fuerza']]

bars = ax.barh(iv_df['variable'][::-1], iv_df['IV'][::-1],
               color=colors_iv[::-1], edgecolor='white', linewidth=1.5, alpha=0.85)
# Líneas de referencia
for x, lab, c in [(0.02, '0.02 (límite inútil)', 'gray'),
                   (0.1,  '0.1 (débil → media)', '#FFD166'),
                   (0.3,  '0.3 (media → fuerte)', '#FF8C42'),
                   (0.5,  '0.5 (fuerte → sospechoso)', '#EF476F')]:
    ax.axvline(x, color=c, linestyle='--', alpha=0.5, linewidth=1)

for bar, iv, fuerza in zip(bars, iv_df['IV'][::-1], iv_df['Fuerza'][::-1]):
    ax.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height()/2,
            f'{iv:.3f} ({fuerza})',
            va='center', color=COLOR_TEXT, fontsize=10, weight='bold')

ax.set_xlabel('Information Value (IV)', size=12)
ax.set_title('Capacidad discriminativa univariante — Information Value', size=13)
ax.grid(axis='x', alpha=0.3, color=COLOR_GRID)
for s in ['top', 'right']: ax.spines[s].set_visible(False)
plt.tight_layout()
save_fb('FB5_iv_woe')
plt.show()


# ============================================================
# CELDA FB-6 — Variable flag MonthlyIncome_missing
# (B.7 del roadmap — esfuerzo mínimo)
# ============================================================
print('🔄 Probando flag MonthlyIncome_missing...\n')

# Esto requiere recargar el dataset original para detectar los NaN antes de imputar
# Si no tienes el original a mano, esta celda es opcional.
# Asumiendo que cargaste df_raw al principio, si no, sustituye por tu DataFrame original
try:
    # Intenta cargar de nuevo el original (ajusta el path si es necesario)
    import pandas as pd
    df_raw = pd.read_csv('cs-training.csv', index_col=0) if os.path.exists('cs-training.csv') else None
    if df_raw is None:
        # Como fallback: la imputación dejó la variable cubierta, pero podemos detectar
        # los antiguos NaN como aquellos con valor exactamente igual a la mediana
        # (heurística aproximada — suficiente para el experimento)
        median_income = df_clean['MonthlyIncome'].median()
        income_was_missing = (df_clean['MonthlyIncome'] == median_income).astype(int)
        print(f'  ⚠ df_raw no disponible, usando heurística: '
              f'{income_was_missing.sum():,} clientes con valor = mediana ({median_income:.0f})')
    else:
        income_was_missing = df_raw['MonthlyIncome'].isna().astype(int)
        print(f'  Detectados {income_was_missing.sum():,} clientes con MonthlyIncome originalmente NaN')

    df_clean['MonthlyIncome_missing'] = income_was_missing.values

    # Reentrenar GB con la variable extra
    features_with_flag = features_model + ['MonthlyIncome_missing']
    X_full2 = df_clean[features_with_flag]
    from sklearn.model_selection import train_test_split
    X_tr2, X_te2, y_tr2, y_te2 = train_test_split(
        X_full2, df_clean[target], test_size=0.25, random_state=42,
        stratify=df_clean[target]
    )
    gb2 = GradientBoostingClassifier(random_state=42)
    gb2.fit(X_tr2, y_tr2)
    auc_with = roc_auc_score(y_te2, gb2.predict_proba(X_te2)[:, 1])
    auc_without = best['auc']

    print(f'\n  AUC sin flag: {auc_without:.4f}')
    print(f'  AUC con flag: {auc_with:.4f}')
    print(f'  Δ AUC: {(auc_with - auc_without)*1000:+.1f} milésimas')
    print(f'  Importancia de MonthlyIncome_missing: {gb2.feature_importances_[-1]:.4f}')
except Exception as e:
    print(f'  ⚠ No ha sido posible reproducir este test: {e}')
    print(f'  Para ejecutarlo: tener el CSV original y crear el flag antes de imputar.')


# ============================================================
# CELDA FB-7 — Tuning hiperparámetros de Gradient Boosting
# (B.6 del roadmap — esfuerzo medio)
# Tarda 2-5 min según hardware. Si no quieres esperar, sáltala.
# ============================================================
print('🔄 GridSearch de hiperparámetros (puede tardar 2-5 min)...\n')

param_grid = {
    'learning_rate': [0.05, 0.1, 0.15],
    'n_estimators':  [100, 200, 300],
    'max_depth':     [3, 4, 5],
}

gb_base = GradientBoostingClassifier(random_state=42, subsample=0.9)
gs = GridSearchCV(gb_base, param_grid, scoring='roc_auc',
                  cv=StratifiedKFold(n_splits=3, shuffle=True, random_state=42),
                  n_jobs=-1, verbose=0)
gs.fit(X_train, y_train)

print(f'\n  Mejores parámetros: {gs.best_params_}')
print(f'  Mejor AUC en CV: {gs.best_score_:.4f}')

# AUC en test del modelo tuneado
gb_tuned = gs.best_estimator_
y_proba_tuned = gb_tuned.predict_proba(X_test)[:, 1]
auc_tuned_test = roc_auc_score(y_test, y_proba_tuned)
print(f'  AUC en test del modelo tuneado: {auc_tuned_test:.4f}')
print(f'  vs. AUC original (defaults): {best["auc"]:.4f}')
print(f'  Δ AUC: {(auc_tuned_test - best["auc"])*1000:+.1f} milésimas')


# ============================================================
# CELDA FB-FINAL — Empaquetar todos los gráficos del feedback
# ============================================================
import zipfile
from google.colab import files

zip_path = 'graficos_feedback.zip'
with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zf:
    for f in sorted(os.listdir(OUT_FB)):
        zf.write(os.path.join(OUT_FB, f), arcname=f)
        print(f'   ✓ {f}')

print(f'\n✅ Zip creado: {zip_path}')
files.download(zip_path)
```

```python
# ============================================================
# REENTRENAMIENTO SIN DATA LEAKAGE (VERSIÓN CORREGIDA)
# ============================================================

from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import roc_auc_score
import numpy as np
import pandas as pd

# 1. Partición TRAIN / TEST (ANTES de cualquier transformación)
X = df.drop(columns=["SeriousDlqin2yrs"])
y = df["SeriousDlqin2yrs"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

# 2. Imputación SOLO con TRAIN
income_median = X_train["MonthlyIncome"].median()

X_train["MonthlyIncome"] = X_train["MonthlyIncome"].fillna(income_median)
X_test["MonthlyIncome"] = X_test["MonthlyIncome"].fillna(income_median)

# Dependents (si lo usas)
if "NumberOfDependents" in X_train.columns:
    X_train["NumberOfDependents"] = X_train["NumberOfDependents"].fillna(0)
    X_test["NumberOfDependents"] = X_test["NumberOfDependents"].fillna(0)

# 3. Winsorización SOLO con TRAIN
def winsorize(series, lower=0.01, upper=0.99):
    low = series.quantile(lower)
    high = series.quantile(upper)
    return np.clip(series, low, high), low, high

# Ajustar en TRAIN
X_train["RevolvingUtilizationOfUnsecuredLines"], rev_low, rev_high = winsorize(
    X_train["RevolvingUtilizationOfUnsecuredLines"]
)

X_train["DebtRatio"], debt_low, debt_high = winsorize(
    X_train["DebtRatio"]
)

# Aplicar en TEST (sin recalcular)
X_test["RevolvingUtilizationOfUnsecuredLines"] = np.clip(
    X_test["RevolvingUtilizationOfUnsecuredLines"], rev_low, rev_high
)

X_test["DebtRatio"] = np.clip(
    X_test["DebtRatio"], debt_low, debt_high
)

# 4. Feature engineering (en ambos)
def feature_engineering(df):
    df = df.copy()

    df["TotalLatePayments"] = (
        df["NumberOfTime30-59DaysPastDueNotWorse"] +
        df["NumberOfTime60-89DaysPastDueNotWorse"] +
        df["NumberOfTimes90DaysLate"]
    )

    df["HasLatePayments"] = (df["TotalLatePayments"] > 0).astype(int)

    df["IncomePerDependent"] = df["MonthlyIncome"] / (df["NumberOfDependents"] + 1)

    return df

X_train = feature_engineering(X_train)
X_test = feature_engineering(X_test)

# 5. Modelo (igual que el tuyo)
model = GradientBoostingClassifier(random_state=42)
model.fit(X_train, y_train)

# 6. Evaluación
y_pred_proba = model.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, y_pred_proba)

print("AUC SIN DATA LEAKAGE:", round(auc, 4))
```

```python
# ============================================================
# REENTRENAMIENTO SIN DATA LEAKAGE (USANDO EL EXCEL SUBIDO)
# ============================================================

from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import roc_auc_score
import numpy as np
import pandas as pd

# Usa el Excel que acabas de cargar
df_raw = df_raw.copy()

# Separar target
X = df_raw.drop(columns=["SeriousDlqin2yrs"]).copy()
y = df_raw["SeriousDlqin2yrs"].copy()

# Split antes de cualquier transformación
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

# ---------- Imputación solo con TRAIN ----------
income_median = X_train["MonthlyIncome"].median()
X_train.loc[:, "MonthlyIncome"] = X_train["MonthlyIncome"].fillna(income_median)
X_test.loc[:, "MonthlyIncome"] = X_test["MonthlyIncome"].fillna(income_median)

if "NumberOfDependents" in X_train.columns:
    X_train.loc[:, "NumberOfDependents"] = X_train["NumberOfDependents"].fillna(0)
    X_test.loc[:, "NumberOfDependents"] = X_test["NumberOfDependents"].fillna(0)

# ---------- Winsorización solo con TRAIN ----------
def winsorize_fit(series, lower=0.01, upper=0.99):
    low = series.quantile(lower)
    high = series.quantile(upper)
    return low, high

def winsorize_apply(series, low, high):
    return np.clip(series, low, high)

rev_low, rev_high = winsorize_fit(X_train["RevolvingUtilizationOfUnsecuredLines"])
debt_low, debt_high = winsorize_fit(X_train["DebtRatio"])

X_train.loc[:, "RevolvingUtilizationOfUnsecuredLines"] = winsorize_apply(
    X_train["RevolvingUtilizationOfUnsecuredLines"], rev_low, rev_high
)
X_test.loc[:, "RevolvingUtilizationOfUnsecuredLines"] = winsorize_apply(
    X_test["RevolvingUtilizationOfUnsecuredLines"], rev_low, rev_high
)

X_train.loc[:, "DebtRatio"] = winsorize_apply(X_train["DebtRatio"], debt_low, debt_high)
X_test.loc[:, "DebtRatio"] = winsorize_apply(X_test["DebtRatio"], debt_low, debt_high)

# ---------- Feature engineering ----------
def feature_engineering(df_in):
    df_out = df_in.copy()

    df_out["TotalLatePayments"] = (
        df_out["NumberOfTime30-59DaysPastDueNotWorse"] +
        df_out["NumberOfTime60-89DaysPastDueNotWorse"] +
        df_out["NumberOfTimes90DaysLate"]
    )

    df_out["HasLatePayments"] = (df_out["TotalLatePayments"] > 0).astype(int)
    df_out["IncomePerDependent"] = df_out["MonthlyIncome"] / (df_out["NumberOfDependents"] + 1)

    return df_out

X_train = feature_engineering(X_train)
X_test = feature_engineering(X_test)

# ---------- Modelo ----------
model = GradientBoostingClassifier(random_state=42)
model.fit(X_train, y_train)

# ---------- Evaluación ----------
y_pred_proba = model.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, y_pred_proba)

print("AUC SIN DATA LEAKAGE:", round(auc, 4))
```
